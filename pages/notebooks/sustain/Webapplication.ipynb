```python
# Enter the Freesurfer volumes of the different regions. The minimum and maximum 
# values permitted are determined by the values encountered in the Rotterdam study

# Note: Adjust the slider based on your values and press shift + Enter
import random
SubjectID = '%010x' % random.randrange(16**10) # Generate a random subject ID for your own records
print('Your randomly generated Subject ID is:',str(SubjectID))
IntracranialCavity = 1483810 #@param {type:"slider", min:994598, max:2171845, step:1}
TotalBrain = 1019869 #@param {type:"slider", min:684784, max:1393898, step:1}
Ventricles = 36248 #@param {type:"slider", min:7163, max:193460, step:1}
Hippocampus = 7639 #@param {type:"slider", min:4019, max:11130, step:1}
Precuneus = 17763 #@param {type:"slider", min:10784, max:26559, step:1}
MiddleTemporal = 19987 #@param {type:"slider", min:11600, max:32870, step:1}
Fusiform = 17262 #@param {type:"slider", min:10833, max:27049, step:1}
Entorhinal = 3621 #@param {type:"slider", min:1449, max:7714, step:1}
FrontalCortex = 105924 #@param {type:"slider", min:69371, max:153637, step:1}
ParietalCortex = 36281 #@param {type:"slider", min:22223, max:51710, step:1}
TemporalCortex = 86936 #@param {type:"slider", min:49413, max:128443, step:1}
OccipitalCortex = 39099 #@param {type:"slider", min:23219, max:60222, step:1}
```

```python
import AgingSubtypeInterface as asi
from AgingSubtypeInterface import predict

Features = [TotalBrain, Ventricles, Hippocampus, Precuneus, MiddleTemporal, Fusiform, Entorhinal, FrontalCortex,
            ParietalCortex, TemporalCortex, OccipitalCortex]
prediction=predict.aging_subtype(Features, IntracranialCavity)
asi.create_report(SubjectID, prediction)
```
